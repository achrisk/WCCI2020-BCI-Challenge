```python
from model_utils import *
from data_utils import *

import numpy as np

from sklearn import svm
from pyriemann.classification import FgMDM
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

import tensorflow as tf
from tensorflow import keras

from sklearn import metrics

import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2
%matplotlib inline

physical_devices = tf.config.experimental.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
sample_rate = 512 #in Hz
num_channels = 12
```

```python
patients = range(1,9)
num_folds = 10
num_epochs = 500
num_hid = 16
flt_size = 128

for i in patients:
    patient = [i]
    dnn_score = holdout(get_model, patients=patient, num_epochs=num_epochs, num_hid=num_hid, mode="inter")
    svm_score = holdout(svm.SVC, patients=patient, mode="inter")
    lda_score = holdout(LinearDiscriminantAnalysis, patients=patient, mode="inter")
    fgmdm_score = holdout(model_fn=FgMDM, data_fn=get_covar, patients=patient, mode="inter")
    eegnet_score = holdout(eegnet, patients=patient, num_epochs=num_epochs, flt_size=flt_size, mode="inter")

    X_train, y_train, X_test, y_test, X_eval = get_data_inter(patient[0], low=8, high=35)
    X_eval_r = get_riemann_ts(X_eval)
    covX_eval = get_covar(X_eval)
    #DNN+SVM+EEGNet
    pred_accum = np.zeros(X_test.shape[0])

    X_train_r = get_riemann_ts(X_train)
    X_test_r = get_riemann_ts(X_test)
    #DNN
    model = get_model(in_shape=X_train_r.shape[-1], num_hidden=16)
    history = model.fit(X_train_r, y_train, epochs=300, verbose=0)
    y_pred = (np.rint(np.squeeze(model.predict(X_test_r)))).astype(int)
    pred_accum += y_pred
    #SVM
    model = svm.SVC()
    model.fit(X_train_r, y_train)
    y_pred = model.predict(X_test_r)
    pred_accum += y_pred
    #EEGNet
    if X_train.ndim == 3:
        X_train = np.expand_dims(X_train, axis=3)
    if X_test.ndim == 3:
        X_test = np.expand_dims(X_test, axis=3)
    if X_eval.ndim == 3:
        X_eval = np.expand_dims(X_eval, axis=3)
    model = eegnet(C1=128)
    history = model.fit(X_train, y_train, epochs=300, verbose=0)
    y_pred = (np.rint(np.squeeze(model.predict(X_test)))).astype(int)
    pred_accum += y_pred
    #Majority Ensemble
    for i in range(X_test.shape[0]):
        if pred_accum[i]>1:
            pred_accum[i] = 1
        else:
            pred_accum[i] = 0
    
    ensemble_score = kappa = metrics.cohen_kappa_score(pred_accum, y_test)
    test_acc = metrics.accuracy_score(pred_accum, y_test)

    labels = ['DNN', 'SVM', 'FgMDM-Coh', 'LDA', 'EEGNet', 'Ensemble\n(DNN+\nSVM+\nEEGNet)']
    x_pos = np.arange(len(labels))
    CTEs = [dnn_score, svm_score, fgmdm_score, lda_score, eegnet_score, ensemble_score]
    fig, ax = plt.subplots()
    ax.bar(x_pos, CTEs,
           align='center',
           alpha=0.5,
           ecolor='black',
           capsize=10)
    ax.set_ylabel('Kappa Score')
    ax.set_xticks(x_pos)
    ax.set_xticklabels(labels)
    ax.set_title('Comparison of Kappa Score for Patient %d (Cross)'%(patient[0]))
    ax.yaxis.grid(True)
    plt.tight_layout()
    plt.savefig('patient%d(cross).png'%(patient[0]))
    plt.show()
```
