```python
from model_utils import *
from data_utils import *

import numpy as np

from sklearn import svm
from pyriemann.classification import FgMDM
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2
%matplotlib inline

physical_devices = tf.config.experimental.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
sample_rate = 512 #in Hz
num_channels = 12
```

```python
patients = range(1,9)
num_folds = 10
num_epochs = 300
num_hid = 16
flt_size = 128

for i in patients:
    patient = [i]
    dnn_scores = cross_val(get_model, patients=patient, num_epochs=num_epochs, num_hid=num_hid)
    svm_scores = cross_val(svm.SVC, patients=patient)
    lda_scores = cross_val(LinearDiscriminantAnalysis, patients=patient)
    fgmdm_scores = cross_val(model_fn=FgMDM, data_fn=get_covar, patients=patient)
    eegnet_scores = cross_val(eegnet, patients=patient, num_epochs=num_epochs, flt_size=flt_size)

    dnn_mean = np.mean(dnn_scores)
    svm_mean = np.mean(svm_scores)
    fgmdm_mean = np.mean(fgmdm_scores)
    lda_mean = np.mean(lda_scores)
    eegnet_mean = np.mean(eegnet_scores)

    dnn_std = np.std(dnn_scores)
    svm_std = np.std(svm_scores)
    fgmdm_std = np.std(fgmdm_scores)
    lda_std = np.std(lda_scores)
    eegnet_std = np.std(eegnet_scores)

    labels = ['DNN', 'SVM', 'FgMDM-Cov', 'LDA', 'EEGNet']
    x_pos = np.arange(len(labels))
    CTEs = [dnn_mean, svm_mean, fgmdm_mean, lda_mean, eegnet_mean]
    error = [dnn_std, svm_std, fgmdm_std, lda_std, eegnet_std]
    fig, ax = plt.subplots()
    ax.bar(x_pos, CTEs,
           yerr=error,
           align='center',
           alpha=0.5,
           ecolor='black',
           capsize=10)
    ax.set_ylabel('Kappa Score')
    ax.set_xticks(x_pos)
    ax.set_xticklabels(labels)
    ax.set_title('Comparison of Kappa Score for Patient %d (Within)'%(patient[0]))
    ax.yaxis.grid(True)
    plt.tight_layout()
    plt.savefig('patient%d(within).png'%(patient[0]))
    plt.show()
```
